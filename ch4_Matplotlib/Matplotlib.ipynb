```python
import matplotlib as mpl, numpy as np, pandas as pd
import matplotlib.pyplot as plt

# 스타일 설정
plt.style.use("classic")
```

```python
# plt.show()
# 이벤트 루프 시작, 현재 활성화된 모든 그림 객채 표시
# 시스템 대화형 그래픽 백엔드와 상호작용
# 파이썬 세션당 한 번만 사용 가능, 대체로 스크립트 맨 마지막에 사용
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

```python
# IPython 셸에서 플로팅
# 모드 - 매직 명령어 %matplotlib
# plt.show() 사용 불 필요

# IPython 노트북에서 플로팅
# %matplotlib notebook -> 대화형 플롯
# %matplotlib inline -> 정적 이미지 (일반적으로 선택)
# -> 노트북 내의 셀에 결과 그래픽 png 이미지 삽입

%matplotlib inline
import numpy as np

x = np.linspace(0, 10, 100)
fig = plt.figure() # figure 생성, 이 후 plot 모두 저장
plt.plot(x, np.sin(x), "-")
plt.plot(x, np.cos(x), "--"); # add ";" in the end to hide text output
```

```python
# 그림을 파일로 저장하기
# .savefig() 명령어
fig.savefig("my_figure.png")
print(fig.canvas.get_supported_filetypes(), "\n") # 시스템 지원 파일 형식

# Image 객체로 파일 내용 표시
from IPython.display import Image
Image("my_figure.png")
```

```python
# MATLAB style interface
# 상태 저장 -> plt 적용되는 곳에 현재 그림과 축을 기록
plt.figure() # 플롯 그림 생성

# 두 개의 패널 중 첫 번째 패널을 생성, 현재 축(axis) 설정
plt.subplot(2, 1, 1) # plt.subplot(rows, columns, panel number)
plt.plot(x, np.sin(x))

# 두 번째 패널을 생성, 현재 축(axis) 설정
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));

# plt.gcf(), plt.gca() - get current figure/axis
```

```python
# Object-Oriented interface
# methods from Figure, Axes Object

# 먼저 플롯 그리드 생성
# ax는 두 개의 축 객체의 배열
fig, ax = plt.subplots(2) # plt.subplots(# of axis)

# 적절한 객체에서 plot() 메서드 호출
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));

# plt.plot() [MATLAB]
# ax.plot() [Ojbect-Oriented]
```
