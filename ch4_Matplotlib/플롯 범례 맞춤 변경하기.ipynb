```python
%matplotlib inline
import matplotlib.pyplot as plt, numpy as np
plt.style.use("classic")
```

```python
# plt.legend()
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), "-b", label="Sine")
ax.plot(x, np.cos(x), "--r", label="Cosine")
ax.axis("equal")
leg = ax.legend();
```

```python
# 범례 테두리를 없애고 그 위치를 지정
ax.legend(loc="upper left", frameon=False)
fig
```

```python
# 범례에 사용되는 열의 개수를 지정 (ncol)
ax.legend(frameon=False, loc="lower center", ncol=2)
fig
```

```python
# 둥근 모서리박스(fancybox)를 테두리로 사용, 음영을 추가, 테두리의 투명도 변경, 텍스트굵기...
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
fig
```

```python
# 범례에 사용할 요소 선택
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
lines = plt.plot(x, y)

# 라인은 plt.line2D 인스턴스의 목록이다
plt.legend(lines[:2], ["first", "second"]);
```

```python
# plt.legend() -> label 속성 집합을 가지지 않은 요소 모두 무시
plt.plot(x, y[:, 0], label="first")
plt.plot(x, y[:, 1], label="second")
plt.plot(x, y[:, 2:])
plt.legend(framealpha=1, frameon=True);
```

```python
# 점 크기에 대한 범례
# 캘리포니아주 도시들의 인구를 표시하는 예제

import pandas as pd
cities = pd.read_csv("california_cities.csv")
# cities.head()

# 관심 있는 데이터 추출
lat, lon = cities["latd"], cities["longd"]
population, area = cities["population_total"], cities["area_total_km2"]

# 레이블은 없지만 크기와 색상을 사용해 점을 뿌림
plt.scatter(lon, lat, label=None,
            c=np.log10(population), cmap="viridis",
            s=area, linewidth=0, alpha=0.5)
plt.axis("equal")
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.colorbar(label="log$_{10}$(population)")
plt.clim(3, 7);

# 범례 생성
# 원하는 크기와 레이블로 빈 리스트를 플로팅
for area in [100, 300, 500]:
    plt.scatter([], [], c="k", alpha=0.3, s=area,
                label=str(area) + " km$^2$")
plt.legend(scatterpoints=1, frameon=False,
           labelspacing=1, title="City Area")

plt.title("California Cities: Area and Population");

# 범례는 항상 플롯상 객체를 참조 -> 빈 리스트 플로팅, 범례 맵핑
```

```python
# 다중 범례
# 아티스트 추가 ax.add_artist()
fig, ax = plt.subplots()

lines = [] # list of plt.line2D instances
styles =["-", "--", "-.", ":"]
x = np.linspace(0, 10, 1000)
for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2), styles[i], color="black")
ax.axis("equal");

# 첫 번째 범례의 라인과 레이블을 지정
ax.legend(lines[:2], ["line A", "line B"],
          loc="upper right", frameon=False)

# 두 번째 범례를 만들고 직접 아티스트를 추가
from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ["line C", "line D"], loc="lower right", frameon=False)
ax.add_artist(leg);
```

```python
# 색상 막대 맞춤 변경하기
# plt.colorbar
x = np.linspace(0, 10, 1000)
I = np.sin(x) * np.cos(x[:, np.newaxis])

plt.imshow(I)
plt.colorbar();
```

```python
# 색상 막대 맞춤 변경하기
plt.imshow(I, cmap="gray");
```

```python
# 색상 지도 선택하기

# 순차적 색상 지도 (순서를 가지는 색상)
# 발산적 색상 지도 (평균을 기준으로 양수 및 음수 편차 구별된 색상)
# 정성적 색상 지도 (순서 없이 색을 혼합)

from matplotlib.colors import LinearSegmentedColormap

def grayscale_cmap(cmap):
    """색상 지도의 회색 조 버전을 반환"""
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))
    
    # RGBA를 감지할 수 있는 회색 조 밝기로 전환
    RGB_weight = [0.299, 0.587, 0.114]
    luminance = np.sqrt(np.dot(colors[:, :3] ** 2, RGB_weight))
    colors[:, :3] = luminance[:, np.newaxis]
    
    return LinearSegmentedColormap.from_list(cmap.name + "_gray", colors, cmap.N)

def view_colormap(cmap):
    """색상 지도를 그에 상응하는 회색 조 버전과 함께 플로팅"""
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))
    cmap = grayscale_cmap(cmap)
    grayscale = cmap(np.arange(cmap.N))
    
    fig, ax = plt.subplots(2, figsize=(6,2),
                           subplot_kw=dict(xticks=[], yticks=[]))
    ax[0].imshow([colors], extent=[0, 10, 0, 1])
    ax[1].imshow([grayscale], extent=[0, 10, 0, 1])
    
view_colormap("jet")
```

```python
view_colormap("viridis")
```

```python
# 색상 한계 및 확장
# 색상 막대 -> plt.Axes의 인스턴스

# 이미지 픽셀의 1% 노이즈 생성
speckles = (np.random.random(I.shape) < 0.01)
I[speckles] = np.random.normal(0, 3, np.count_nonzero(speckles))

plt.figure(figsize=(10, 3.5))

plt.subplot(1, 2, 1)
plt.imshow(I, cmap="RdBu")
plt.colorbar()

plt.subplot(1, 2, 2)
plt.imshow(I, cmap="RdBu")
plt.colorbar(extend="both")
plt.clim(-1, 1);
```

```python
# 불연속적인 색상 막대
# plt.cm.get_cmap(cmap, # of bins)
plt.imshow(I, cmap=plt.cm.get_cmap("Blues", 6))
plt.colorbar()
plt.clim(-1, 1);
```

```python
# 예제: 손으로 쓴 숫자
# 0-5까지 숫자의 이미지를 적재하고 그중 몇 가지를 시각화
from sklearn.datasets import load_digits
digits = load_digits(n_class=6)

# 각 숫자가 64픽셀의 색조로 정의, 각 숫자를 64차원 공간에 놓인 점으로 생각
fig, ax = plt.subplots(8, 8, figsize=(6,6))

for i, axi in enumerate(ax.flat):
    axi.imshow(digits.images[i], cmap="binary")
    axi.set(xticks=[], yticks=[]) # passing empty x/y-ticks remove all tick marks
```

```python
# IsoMap을 사용해 숫자를 2차원에 사영
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
projection = iso.fit_transform(digits.data)

# 결과 플롯 그리기
plt.scatter(projection[:, 0], projection[:, 1], lw=0.1,
            c=digits.target, cmap=plt.cm.get_cmap("cubehelix", 6))
plt.colorbar(ticks=range(6), label="digit value")
plt.clim(-0.5, 0.5)
```
