```python
%matplotlib inline
import matplotlib.pyplot as plt, numpy as np
plt.style.use("seaborn-whitegrid")
```

```python
# plt.plot scatter plot
x = np.linspace(0, 10, 30)
y = np.sin(x)

# 표시자 (marker)
plt.plot(x, y, "o", color="k");
```

```python
rng = np.random.RandomState(0)
for marker in ["o",".",",","x","+","v","^","<",">","s","d"]:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker=\"{0}\"".format(marker))
    plt.legend(numpoints=1)
    plt.xlim(0, 1.8);
```

```python
# 문자 코드를 선과 색상 코드와 함께 사용
plt.plot(x, y, "-ok"); # 선(-), 원 표시 기호(o), 검정색(k)
```

```python
plt.plot(x, y, "-p", color="gray",
         markersize=15, linewidth=4,
         markerfacecolor="white",
         markeredgecolor="gray",
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

```python
# plt.scatter 산점도
# plt.scatter -> 각 점의 속성을 개별적으로 제어, 데이터에 매핑할 수 있는 산점도
plt.scatter(x, y, marker="o");
```

```python
# 다양한 색상과 크기를 갖는 점으로 이뤄진 임의의 산점도
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap="viridis")
plt.colorbar(); # 색상 척도 표시
```

```python
# 붓꽃 데이터, 세가지 유형의 꽃 중 하나로 그 꽃잎과 꽃받침의 크기를 세밀하게 측정한 값
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

# x,y = sepal length, width
# marker size, color = petal width, species
plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap="viridis")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```
